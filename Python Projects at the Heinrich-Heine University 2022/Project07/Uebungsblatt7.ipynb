Das folgende Programm soll eigentlich die Häufigkeit derjenigen Trigramme aus ``grimm.txt`` ausgeben, die <b>fünfmal oder öfter</b> vorkommen. Hierbei soll die Groß- und Kleinschreibung der Wörter und Sonderzeichen ignoriert werden, und die Ausgabe nach der Häufigkeit sortiert erfolgen:

```python
f = open("grimm.txt", encoding="utf-8")

    key = (word, nextword, nextnextword)
    trigrams[key] = trigrams.get(key, 0) + 1   
        
l = list(trigrams.items())

```

Allerdings hat das Programm einen **Fehler**: es stürzt am Ende der Datei ab.

Kopieren Sie das Programm in die Antwortzelle und beheben Sie den Fehler, aber lassen Sie möglichst große Teile des Programs intakt. Es reicht aus, eine Zeile des Programms zu ändern.

````python
f = open("grimm.txt", encoding="utf-8")
text = f.read()
f.close()

words = []

for word in text.split():
    word = word.strip(":,.!?-–»«;")
    if len(word) != 0:
        words.append(word)

trigrams = {}
for i in range(len(words)):
    word = words[i-2]
    nextword = words[i-1]
    nextnextword = words[i]
    key = (word, nextword, nextnextword)
    trigrams[key] = trigrams.get(key, 0) + 1   

l = list(trigrams.items())
l.sort(key=lambda x: x[1])
l.reverse()

for key, count in l:
    if count < 5:
        break
    word = key[0]
    nextword = key[1]
    nextnextword = key[2]
    print(word, nextword, nextnextword, count)
````

Der Inhalt der Datei ist eine Liste aus Dictionaries, ein Dictionary pro Schule. Die Schlüssel der Dictionaries sind für alle Schulen gleich. Schreiben Sie ein Programm, das die Anzahl der **Grundschulen** in NRW pro Postleitzahl zeilenweise ausgibt, sortiert nach Postleitzahl. <b>Es sollen nur die Postleitzahlen ausgegeben werden, welche mindestens eine Grundschule beheimaten.</b> Sie können den folgenden Code verwenden, um die JSON-Datei zu lesen:

Die ersten Zeilen der Ausgabe sind:

**Hinweis**: Sie können den Code von Blatt 6 Aufgabe 1 wiederverwenden!

**Hinweis**: Verwenden Sie ein Dictionary, das als Schlüssel die Postleitzahlen und als Werte die Anzahlen der Grundschulen mit der jeweiligen Postleitzahl speichert.

````python
import json
with open("nrw-schulen.json", encoding="utf-8") as f:
    data = json.loads(f.read())
    
#print(data)    

lst = [ ]
for d in data:
    if d["school_type"] == "Grundschule":
        grund = (d["address"].split(", ")[-1])
        lst.append(grund)

lst.sort()  


result = dict((i, lst.count(i)) for i in lst)
#print (result)

for key, value in result.items() :
    print (key, value)
#print(lst)
````
