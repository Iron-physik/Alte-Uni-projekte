## Aufgabe 1 (2 Punkte)
Schreiben Sie ein Programm, welches die Quadratzahlen der Zahlen 0 bis 100 einschließlich ausgibt, eine pro Zeile.

Die ersten Quadratzahlen lauten ``0, 1, 4, 9, 16``; die letzte, die auszugeben ist, lautet ``10000``.

````python
x = range(0, 101, 1)
for zahl in x:
    quadratzahl=zahl*zahl
    print(quadratzahl) 
````

## Aufgabe 2 (4 Punkte)
Schreiben Sie ein Programm, dass die Zahlen von 1 bis 100 (inklusive) zeilenweise ausgibt. Zahlen, die durch drei teilbar sind, sollen durch das Wort ``"Fizz"`` ersetzt werden; Zahlen die durch fünf teilbar sind, sollen durch das Wort ``"Buzz"`` ersetzt werden; und Zahlen, die durch drei *und* fünf teilbar sind sollen durch ``"FizzBuzz"`` ersetzt werden. Die ersten paar Zeilen der Ausgabe sind also:

```
1
2
Fizz
4
Buzz
Fizz
7
8
Fizz
Buzz
11
Fizz
13
14
FizzBuzz
16
```

Hilfestellung: um herauszufinden, ob eine Zahl ``a`` durch drei teilbar ist, können sie die [Modulo-Operation](http://stackoverflow.com/a/4432235) benutzen: ``a % 3 == 0``.

Programme von diesem Schwierigkeitsgrad [werden](https://imranontech.com/2007/01/24/using-fizzbuzz-to-find-developers-who-grok-coding/) (unsinnigerweise) von einigen Firmen in Vorstellungsgesprächen benutzt, um BewerberInnen auszusortieren.

````python
x = range(1, 101, 1)
for zahl in x:
    if zahl%5==0 and zahl%3==0: 
        print("FizzBuzz")
    elif zahl%5==0:
        print("Buzz")
    elif zahl%3==0:
        print("Fizz")
    else:
        print(zahl)
````
