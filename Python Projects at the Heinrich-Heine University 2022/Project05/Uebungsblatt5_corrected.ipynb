## Aufgabe 2 (4 Punkte)

Schreiben Sie ein Programm, das Zahlen in der Datei ``numbers.txt`` einliest und sie nach numerisch absteigender Reihenfolge sortiert. Danach sollen sie letzten 50 Zahlen (d.h. die 50 kleinsten) entfernt werden. Anschließen soll das Programm die Zahlen zeilenweise ausgeben.

**Hinweis**: um einen String in eine Kommazahl umzuwandeln, verwenden Sie die Funtion ``float``. Zum Beispiel ist ``float("4.2") == 4.2``.

<b>4 von 4 Punkten:</b>


````python
f = open("numbers.txt", encoding="utf-8")
content = f.read()
f.close()

numbers = [float(i) for i in content.split()]
#print(type(numbers))
#print (numbers)
#numbers = [float(item) for item in content]

numbers.sort(reverse=True)

del numbers[-50:]

numbers = [str(x) for x in numbers]

for item in numbers:
    print(item)
````
